# Part of Speech in NLP 

```python
# We will see how to work POS in nlp 

sent = " My name is Ankita Shendge"
sent_tokens = word_tokenize(sent)
sent_tokens
```

```python
for token in sent_tokens:
    print(nltk.pos_tag([token]))
```

# Name Entity Recognization in NLP 

```python
from nltk import ne_chunk
```

```python
NE_sent = ' The US president stays in the WHITEHOUSE'
```

```python
NE_tokens = word_tokenize(NE_sent)
# after tokenize need to add the pos tags 
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
NE_tags
```

```python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
```
